```python
sample_data = """be cfbegad cbdgef fgaecd cgeb fdcge agebfd fecdb fabcd edb | fdgacbe cefdb cefbgd gcbe
edbfga begcd cbg gc gcadebf fbgde acbgfd abcde gfcbed gfec | fcgedb cgb dgebacf gc
fgaebd cg bdaec gdafb agbcfd gdcbef bgcad gfac gcb cdgabef | cg cg fdcagb cbg
fbegcd cbd adcefb dageb afcb bc aefdc ecdab fgdeca fcdbega | efabcd cedba gadfec cb
aecbfdg fbg gf bafeg dbefa fcge gcbea fcaegb dgceab fcbdga | gecf egdcabf bgf bfgea
fgeab ca afcebg bdacfeg cfaedg gcfdb baec bfadeg bafgc acf | gebdcfa ecba ca fadegcb
dbcfg fgd bdegcaf fgec aegbdf ecdfab fbedc dacgb gdcebf gf | cefg dcbef fcge gbcadfe
bdfegc cbegaf gecbf dfcage bdacg ed bedf ced adcbefg gebcd | ed bcgafe cdgba cbgef
egadfb cdbfeg cegd fecab cgb gbdefca cg fgcdab egfdb bfceg | gbdfcae bgc cg cgb
gcafb gcf dcaebfg ecagb gf abcdeg gaef cafbge fdbac fegbdc | fgae cfgab fg bagce"""

sample_data = open('d8.input').read()
{
    "1": 2,
    "4": 4,
    "7": 3,
    "8": 7,

    "2": 5,
    "3": 5,
    "5": 5,
    
    "0": 6,
    "6": 6,
    "9": 6
}

def deduce(decoder):
    results = {}
    segments = {}
    decoder_len = {d: len(d) for d in decoder.split()}
    fives = []
    sixes = []
    rights = []
    tl_and_middle = []
    for d, l in decoder_len.items():
        if l == 2:
            results[1] = d
        if l == 3:
            results[7] = d
        if l == 4:
            results[4] = d
        if l == 7:
            results[8] = d
        if l == 5:
            fives.append(d)
        if l == 6:
            sixes.append(d)
    for s in results[7]:
        if s not in results[1]:
            segments[s] = SEG.T
        else:
            rights.append(s)
    for s in results[4]:
        if s not in results[1]:
            tl_and_middle.append(s)
    tl, middle = tl_and_middle
    for five in fives:
        if tl in five and middle in five:
            results[5] = five
    fives.remove(results[5])
    tr, br = rights
    for five in fives:
        if tr in five and br in five:
            results[3] = five
    fives.remove(results[3])
    results[2] = fives[0]
    for s in results[5]:
        if s in results[1]:
            segments[s] = SEG.BR
            for right in rights:
                if right != s:
                    segments[right] = SEG.TR
    for s in results[5]:
        if s not in segments and s not in tl_and_middle:
            segments[s] = SEG.B
    for s in results[3]:
        if s not in segments:
            segments[s] = SEG.M
    reverse_seg = { v:k for k,v in segments.items()}
    for six in sixes:
        if reverse_seg[SEG.M] not in six:
            results[0] = six
    sixes.remove(results[0])
    for six in sixes:
        if tr in six and br in six:
            results[9] = six
    sixes.remove(results[9])
    results[6] = sixes[0]
    return {''.join(sorted(v)):k for k,v in results.items()}



import re
total = 0
results = []
for line in sample_data.splitlines():
    decoder, message = line.split(" | ")
    message = [''.join(sorted(m)) for m in message.split()]
    decoder = deduce(decoder)
    results.append(int("".join([str(decoder[m]) for m in message]), 10))
print(sum(results))
```
