```python
sample_data = """([]{<[]((({}[])))>})"""

sample_data = "[(()[<>])]({[<{<<[]>>("

sample_data = """[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]"""

sample_data = open('d10.input').read()

data = sample_data.splitlines()

openers = ['(', '[', '{', '<']
closers = [')', ']', '}', '>']
expected = {'(': ')', '[': ']', '{': '}', '<': '>'}
points = {')': 3, ']': 57, '}': 1197, '>': 25137}

class InvalidGroups(Exception): 
    def __init__(self, *args, index=None, char=None, invalid=None, score=None, **kwargs,):
        self.index = index
        self.char = char
        self.invalid = invalid
        super().__init__(*args, **kwargs)
    
    @property
    def score(self):
        return points[self.invalid]

class IncompleteGroups(Exception):
    def __init__(self, *args, index=None, char=None, **kwargs,):
        self.index = index
        self.char = char
        super().__init__(*args, **kwargs)

DEBUG = True

def pull_chunks(chunk: str) -> int:
    for i in range(16):
        DEBUG and print(f'{i:x}', end="")
    DEBUG and print("")
    DEBUG and print(sample_data)
    class Recurser:
        def __init__(self, chunk):
            self.closed = False
            self.missing = ""
            self.chunk = chunk

        def recurse(self, index, depth):
            # Because a class is too much work
            pad = " " * depth
            depth+=1
            
            #What are we dealing with?
            c = (self.chunk + self.missing)[index]
            DEBUG and print (pad + f"Processing: {index}: {c}")
            if c not in openers:
                DEBUG and print(pad + f"Closer: {(self.chunk + self.missing)[index]}.  Returning Index")
                return index

            #Where we called with nowhere to go?
            if index + 1 == len((self.chunk + self.missing)):
                if c in openers:
                    DEBUG and print(pad + f"End of the line and {c} is missing a {expected[c]}")
                    self.missing += expected[c]


            #Let's ask ourselves what we should do about the next index
            result = self.recurse(index+1, depth)
            DEBUG and print (pad + f"{index}: {c} got result {result}: {(self.chunk + self.missing)[result]}")

            # If we're closed and still here, there's unfinished business
            if self.closed:
                DEBUG and print(f"We're missing a {expected[c]}")
                self.missing += expected[c]
                return self.recurse(result+1, depth)
            
            #If we got what we're looking for
            if (self.chunk + self.missing)[result] == expected[c]:

                # Close
                if result + 1 == len((self.chunk + self.missing)):
                    self.closed = True
                    DEBUG and print(pad + f"Closed expectedly")
                    return result
                DEBUG and print (pad + f"Recursing to {result+1}")

                # Or ask ourselves about the next thing
                return self.recurse(result+1, depth)


            raise InvalidGroups(f"Expected {expected[c]} at index {result}.  Got {(self.chunk + missing)[result]} instead", index=index, char=c, invalid=(self.chunk + missing)[result])
    
    recurser = Recurser(chunk)
    recurser.recurse(0,0)
    return recurser.missing
    
DEBUG = False


def score_missing(missing):
    score = 0
    points = {
        ")": 1,
        "]": 2,
        "}": 3,
        ">": 4,
    }
    elements = list(reversed(missing))
    while elements:
        score = score * 5
        score += points[elements.pop()]
    return score
    


score = 0
missings = []
for line in data:
    missing = ""
    try:
        missing = pull_chunks(line)
        missings.append(score_missing(missing))
        
    except InvalidGroups as exc:
        print(f'Exception!: {line} Missing: {missing} Invalid: {exc.invalid}')
        score += exc.score
    print(f'Line: {line} Missing: {missing}')
print (sorted(missings)[len(missings)//2])
print(score)
    
```
